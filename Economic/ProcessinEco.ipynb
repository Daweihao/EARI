```python
import pandas as pd
```

```python
file1 = "DP03_economic(F).xlsx"
xl = pd.ExcelFile(file1)
xl.sheet_names
eco_pd = xl.parse('MA')
eco_pd = eco_pd.drop(eco_pd.index[0])
eco_pd = eco_pd.drop(['GEO.id'],axis =1)
eco_pd['GEO.id2'] = eco_pd['GEO.id2'].apply(lambda x : int(x /1000000))
eco_pd['HC01_VC85'] = eco_pd['HC01_VC85'].replace('-',0)
eco_pd
```

```python
file2 = "DP03_economic_county.xlsx"
xl2 = pd.ExcelFile(file2)
xl2.sheet_names
county_eco = xl2.parse('MA')
county_eco = county_eco.drop(county_eco.index[0])
median_income = county_eco[['GEO.id2','GEO.display-label','HC01_VC85']]
median_income
```

```python
flags = []
for index1,each in median_income.iterrows():
    for index2,tract in eco_pd.iterrows():
#         print(tract['HC01_VC85'])
        
        if each['GEO.id2'] == tract['GEO.id2']:
            if int(each['HC01_VC85']) < int(tract['HC01_VC85']):
                flags.append(1)
            else:
                flags.append(0)
eco_pd['flag'] = flags                
eco_pd
```

```python
eco_pd.groupby(eco_pd['GEO.id2']).mean().flag
```

```python
eco_pd.to_csv("outcome.csv",index=False,sep=',')
```
